```python
# ðŸ§  Stroke CT Scan EDA

import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from glob import glob
from collections import defaultdict
from PIL import Image
import pathlib
```

```python

%matplotlib inline
sns.set(style='whitegrid')
```

```python
root_dir = os.path.join(os.getcwd(), "Data")

print("Resolved path:", root_dir)

classes = ['Bleeding', 'Ischemia', 'Normal']
```

```python

image_paths = defaultdict(list)
for cls in classes:
    path = os.path.join(root_dir, cls, 'PNG')
    image_paths[cls] = glob(os.path.join(path, '*.png'))
```

```python

# === Count Images ===
for cls in classes:
    print(f"{cls}: {len(image_paths[cls])} images")
```

```python
# === Plot Class Distribution ===
counts = {cls: len(image_paths[cls]) for cls in classes}
sns.barplot(x=list(counts.keys()), y=list(counts.values()))
plt.title("Class Distribution")
plt.ylabel("Number of images")
plt.xlabel("Class")
plt.show()
```

```python
# === Visualize Samples ===
fig, axes = plt.subplots(1, 5, figsize=(15, 4))
cls = 'Bleeding'
for i, path in enumerate(image_paths[cls][:5]):
    img = Image.open(path)
    axes[i].imshow(img, cmap='gray')
    axes[i].axis('off')
    axes[i].set_title(cls)
plt.tight_layout()
plt.suptitle('Sample Bleeding Images', y=1.05)
plt.show()
```
